```python
# Set up Stable Diffusion in Google Colab with Hugging Face Diffusers Library

# 1. Install necessary libraries
!pip install diffusers transformers scipy
!pip install accelerate

# 2. Import libraries and load model
from diffusers import StableDiffusionPipeline
import torch
import os

# Set up your Hugging Face token (necessary to use the model)
# Sign up on Hugging Face (https://huggingface.co/) if you haven't, and get your access token from your account settings
!huggingface-cli login

# 3. Load the model pipeline
model_id = "CompVis/stable-diffusion-v1-4"  # Stable Diffusion model ID
pipe = StableDiffusionPipeline.from_pretrained(model_id)
pipe = pipe.to("cpu")  # use CPU for processing

# 4. Set up the base output directory
output_dir = "/content/drive/MyDrive/Colab Notebooks/Final_Project/121124_pic_2"
os.makedirs(output_dir, exist_ok=True)

# 5. List of items to generate images for (without "png" suffix)
items = [
    # Fruits

    "Group of Tomatoes", "Group of Cucumbers", "Group of White Onions", "Group of Garlic Bulbs", "Group of Yellow Potatoes",
    "Group of Sweet Potatoes", "Group of Zucchinis", "Group of Champignon Mushrooms", "Lettuce", "Group of Avocados",

    # General Items
    "Dry Rice", "Dry Penne Pasta", "White Bread", "White Eggs Tray", "Milk inside Glass Bottle",
    "Sliced Yellow Cheese", "Yellow Butter", "Fresh Chicken Breast", "Fresh Ground Beef",
    "Fresh Fillet Salmon", "glass cooking oil bottle",
    "Oats", "Peanut butter jar", "chocolate tablet"
]

# 6. Generate 15 images for each item and save them in a separate folder
for item in items:
    # Create a sanitized folder name for each item
    item_sanitized = item.replace(" ", "_").replace(",", "")
    item_dir = os.path.join(output_dir, item_sanitized)
    os.makedirs(item_dir, exist_ok=True)  # Create directory for the current item

    # Define the prompt without "png" in the description
    prompt = f"A fridge with only {item} inside, high detail"

    for i in range(30):
        print(f"Generating image {i+1} for item: {item}")

        # Generate image
        image = pipe(prompt).images[0]  # Get the first generated image

        # Save the image with a unique name in the specific item folder
        image_path = os.path.join(item_dir, f"{item_sanitized}_{i+1}.png")
        image.save(image_path)

print(f"Images are saved in {output_dir} with separate folders for each item.")
```

```python
!pip install pytorch-lightning torch torchvision torchmetrics pillow numpy
import pytorch_lightning as L
import torch
from torch import nn, optim
import torch.nn.functional as F
from torchvision import models, transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, random_split
from torchmetrics import Accuracy
from PIL import Image

# Define the transfer learning model with added accuracy metrics
class ImagenetTransferLearning(L.LightningModule):
    def __init__(self):
        super().__init__()

        # Initialize a pre-trained ResNet-50 model
        backbone = models.resnet50(pretrained=True)

        # Extract number of features
        num_filters = backbone.fc.in_features
        layers = list(backbone.children())[:-1]  # Remove the last layer
        self.feature_extractor = nn.Sequential(*layers)

        # Freeze feature extractor layers
        for param in self.feature_extractor.parameters():
            param.requires_grad = False
        self.feature_extractor.eval()

        # Classifier for 39 classes
        num_target_classes = 39
        self.classifier = nn.Linear(num_filters, num_target_classes)

        # Initialize accuracy metrics for multiclass classification with 39 classes
        self.train_accuracy = Accuracy(task='multiclass', num_classes=num_target_classes)
        self.val_accuracy = Accuracy(task='multiclass', num_classes=num_target_classes)

    def forward(self, x):
        representations = self.feature_extractor(x).flatten(1)
        x = self.classifier(representations)
        return x

    def training_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        loss = F.cross_entropy(logits, y)
        acc = self.train_accuracy(logits, y)

        # Log loss and accuracy
        self.log('train_loss', loss, prog_bar=True, on_step=True)
        self.log('train_acc', acc, prog_bar=True, on_step=True)
        return loss

    def validation_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        loss = F.cross_entropy(logits, y)
        acc = self.val_accuracy(logits, y)

        # Log validation loss and accuracy
        self.log('val_loss', loss, prog_bar=True, on_epoch=True)
        self.log('val_acc', acc, prog_bar=True, on_epoch=True)
        return loss

    def configure_optimizers(self):
        return optim.Adam(self.parameters(), lr=1e-3)

# Data transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

# Load custom dataset using ImageFolder
data_dir = '/content/stable_diffusion'  # Update this path to the root directory of your 39 class folders
dataset = ImageFolder(root=data_dir, transform=transform)

# Define 5% of the original data for training and validation
total_samples = len(dataset)
train_samples = int(total_samples * 0.8)  # 80% for training
val_samples = total_samples - train_samples  # 20% for validation

# Split the dataset: 80% for training, 20% for validation
train_subset, val_subset = random_split(dataset, [train_samples, val_samples])

# Create data loaders
train_loader = DataLoader(train_subset, batch_size=3, shuffle=True)
val_loader = DataLoader(val_subset, batch_size=3)

# Initialize model and trainer
model = ImagenetTransferLearning()

trainer = L.Trainer(
    limit_train_batches=1000,  # Data to use each epoch
    max_epochs=10,
)

# Train the model
trainer.fit(model=model, train_dataloaders=train_loader, val_dataloaders=val_loader)

# Function to test the model with a custom image
def classify_image(model, image_path, transform, top_k=10):
    model.eval()  # Set the model to evaluation mode
    image = Image.open(image_path).convert('RGB')  # Load image and convert to RGB
    image = transform(image).unsqueeze(0)  # Apply transforms and add batch dimension

    with torch.no_grad():  # Disable gradient calculation for inference
        logits = model(image)
        probabilities = torch.softmax(logits, dim=1)  # Convert logits to probabilities

        # Get top K probabilities and class indices (set top_k to None to get all classes)
        top_k = top_k if top_k is not None else probabilities.size(1)
        top_probs, top_indices = torch.topk(probabilities, k=top_k, dim=1)

    # Convert probabilities and indices to lists for readability
    top_probs = top_probs.squeeze().tolist()
    top_indices = top_indices.squeeze().tolist()

    # Map class indices to class names
    class_probabilities = [(dataset.classes[idx], prob) for idx, prob in zip(top_indices, top_probs)]

    # Print each class with its associated probability
    print("Class probabilities from highest to lowest:")
    for class_name, prob in class_probabilities:
        print(f"{class_name}: {prob:.4f}")

    return class_probabilities
```

```python

# Example usage after training
# Replace 'path_to_yourbr_image.jpg' with the actual path of an image to classify
classify_image(model, "KALE.jfif", transform)
```
